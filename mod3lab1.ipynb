Section 1: distance matrics

```python
import numpy as np
import pandas as pd
import math
from scipy.spatial import distance
import matplotlib.pyplot as plt
```

i. euclidean distance

in 2D

```python
x_1 = np.array((10, 19))
x_2 = np.array((36, 56))
sumv = 0
for i in range(len(x_1)):
  sumv += ((x_1[i]-x_2[i])**2)
dist = math.sqrt(sumv)
print(dist)
```

vectorised

```python
euclidean_dist = np.sqrt(np.sum((x_1-x_2)**2))
print(euclidean_dist)
```

3D

```python
plotx = np.linspace(-5,10,20)
meshx, meshy = np.meshgrid(plotx,plotx)
tempeuclid = np.sqrt((meshx**2 + meshy**2))
import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(z=tempeuclid, x=plotx, y=plotx)])
fig.update_layout(title='Euclidean Distance from origin', autosize=False,
                  width=1000, height=1000,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```
